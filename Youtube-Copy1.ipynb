```python
data = []
currentpagesource=youtube.get_page_source()
lastpagesource=''

while(True):
    if(lastpagesource==currentpagesource):
        break
        
    lastpagesource=currentpagesource
    response=youtube.video_comments()

    for c in response['body']:
        data.append(c)
        
    youtube.scroll()
    currentpagesource=youtube.get_page_source()
```
